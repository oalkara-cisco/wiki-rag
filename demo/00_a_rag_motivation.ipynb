# RAG  - LangChain - BedRock

## Limitations of LLMs
LLMs are great but they have some limitations.

* Lack of recent knowlege because of the training cut-off date.
* No visibility to internal/proprietary knowledge base.
* To get the right answer, you have ask precise questions.

## RAG motivation

Empowers LLMs to use external knowledge.

### About
* Retrieval-Augmented Generation.
* Combines retrieval and generation processes to enhance the capabilities of LLMs.
* Answer the questions precisely (better than search results)
